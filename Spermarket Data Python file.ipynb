what time of the day do we have a higher sale volume

```python
salesvolumes=mydata["Quantity"]
salesvolumeslist=list(salesvolumes)

time=mydata["Time"]
timelist=list(time)

morningsales=[]
midmorningsales=[]
afternoonsales=[]
eveningsales=[]

timelist=[time.replace(":",".") for time in timelist]


newtimelist=[float (t) for t in timelist]  


finaltimelist=[math.floor(num) for num in newtimelist]
combined=list(zip(finaltimelist,salesvolumeslist))
combined

for t ,s in combined:
    if t>0 and t<6:
        morningsales.append(s)
    if t>=6 and t<12:
        midmorningsales.append(s)
    if t>=12 and t<18:
        afternoonsales.append(s)
    if t>=18 and t<=23:
        eveningsales.append(s)

highestsales=max(sum(morningsales),sum(midmorningsales),sum(afternoonsales),sum(eveningsales)) 

print("Total Morning Sales 00-7:59am:",sum(morningsales))
print("Total Mid-Morning Sales 6am-11:59noon:", sum(midmorningsales))
print("Total Afternoon Sales:12noon-5:59pm:", sum(afternoonsales))
print("Total Evening Sales:6pm-11:59pm:", sum(eveningsales))
print("Time Period with the Highest Sales:", highestsales)        
      
```

```python
#what time of the day do we have a higher sale volume
#Time of day when there are most sales in terms of quantity
timelist
Newtimelist = []
#replacing : with . in the list so that we can be able to remove values after the decimal
for x in timelist:
    mytime = x.replace(":", ".")
    #print(mytime)
    #float is number that carries decimal points. We use float to convert numbers with decimal points to whole numbers
    y = float(mytime)
    finaltime = math.floor(y)
    
    Newtimelist.append(finaltime)

frequencylist = list(zip(salesvolumeslist, newtimelist))

my8to10amlist = []
my10to12pmlist = []
my12to14pmlist = []
my14to16pmlist = []
my16to18pmlist = []
my18to20pmlist = []
my20to22pmlist = []
my22to24pmlist = []
my00to8amlist=[]
for qnty,time in frequencylist:
    if time >=0 and time <=8:
        #print(qnty)
        my00to8amlist.append(qnty)
    if time >8 and time <=10:
        #print(qnty)
        my8to10amlist.append(qnty)
    if time >10 and time <=12:
        #print(qnty)
        my10to12pmlist.append(qnty)
    if time >12 and time <=14:
        #print(qnty)
        my12to14pmlist.append(qnty)
    if time >14 and time <=16:
        #print(qnty)
        my14to16pmlist.append(qnty)
    if time >16 and time <=18:
        #print(qnty)
        my16to18pmlist.append(qnty)
    if time >18 and time <=20:
        #print(qnty)
        my18to20pmlist.append(qnty)
    if time >20 and time <=22:
        #print(qnty)
        my20to22pmlist.append(qnty)
    if time >22 and time <=24:
        #print(qnty)
        my22to24pmlist.append(qnty)

print("00am-8am", sum(my00to8amlist))        
print("8am-10am", sum(my8to10amlist))
print("10am-12pm", sum(my10to12pmlist))
print("12am-14pm", sum(my12to14pmlist))
print("14am-16pm", sum(my14to16pmlist))
print("16am-18pm", sum(my16to18pmlist))
print("18am-20pm", sum(my18to20pmlist))
print("20am-22pm", sum(my20to22pmlist))
print("22am-24pm", sum(my22to24pmlist))  
```

most profitable month

```python
grossincome=mydata["gross income"]
grossincomelist=list(grossincome)

newmonthlist=[]
januarylist=[]
februarylist=[]
marchlist=[]

month=mydata["Date"]
monthlist=list(month)

for m in monthlist:
    newmonthlist.append(m[0])
    
profitmonthlist=list(zip(newmonthlist,grossincomelist))
profitmonthlist

for month,gross_income in profitmonthlist:
    if month=="1":
        januarylist.append(gross_income)
    if month=="2":
        februarylist.append(gross_income)
    if month=="3":
        marchlist.append(gross_income) 
        
highestsales=max(sum(januarylist),sum(februarylist),sum(marchlist))

print("January sales:",sum(januarylist))
print("February sales:",sum(februarylist))
print("March sales:",sum(marchlist))
print("highestsales:",highestsales)

if highestsales==sum(januarylist):
    print("January sales were the highest:",highestsales)
    
if highestsales==sum(februarylist):
    print("February sales were the highest:",highestsales)

if highestsales==sum(marchlist):
    print("March sales were the highest:",highestsales)    
```
